```python
import pandas as pd

# Cargar los datos desde el archivo CSV
file_path = '../files/input/tbl1.csv'

# Definir nombres de columnas y recargar los datos
column_names = ['K0', 'K1', 'c12', 'c13', 'c14', 'c15', 'c16']
data = pd.read_csv(file_path, names=column_names, header=None)
```

```python
# Sumar los valores en la columna 'c12' nuevamente
sum_c12 = data['c12'].sum()
sum_c12
```

```python
# Contar la cantidad de registros en el DataFrame
count_records = data.shape[0]
count_records
```

```python
# Convertir la columna 'c14' a formato de fecha y ordenar los datos por esta columna
data['c14'] = pd.to_datetime(data['c14'])
sorted_data = data.sort_values(by='c14')

# Seleccionar los primeros cinco registros
first_five_records = sorted_data.head()
first_five_records
```

```python
# Filtrar los registros donde la primera letra de 'c16' es igual a 'K0'
filtered_data = data[data.apply(lambda row: row['c16'].startswith(row['K0']), axis=1)]

# Seleccionar solo las columnas 'K0' y 'c16'
selected_columns = filtered_data[['K0', 'c16']]
selected_columns
```

```python
# Cargar los datos desde el nuevo archivo CSV
file_path_tbl0 = '../files/input/tbl0.csv'

# Definir nombres de columnas y recargar los datos para tbl0
column_names_tbl0 = ['K0', 'c01', 'c02', 'c03', 'c04']
data_tbl0 = pd.read_csv(file_path_tbl0, names=column_names_tbl0, header=None)
```

```python
# Filtrar los registros donde el campo 'c02' es igual a 100 o 600
filtered_tbl0 = data_tbl0[(data_tbl0['c02'] == 100) | (data_tbl0['c02'] == 600)]

filtered_tbl0
```

```python
# Filtrar los registros de la tabla 'tbl1' donde 'K0' es igual a 'A' y ordenar por fecha ('c14')
filtered_sorted_tbl1 = data[data['K0'] == 'A'].sort_values(by='c14')

filtered_sorted_tbl1
```

```python
# Filtrar los registros de la tabla 'tbl1' donde 'K0' no es 'A' ni 'B', y 'c13' no es 200 ni 900
# Luego, ordenar por fecha ('c14')
filtered_sorted_tbl1 = data[(~data['K0'].isin(['A', 'B'])) & (~data['c13'].isin([200, 900]))].sort_values(by='c14')

filtered_sorted_tbl1
```

```python
# Cargar los datos desde el archivo CSV para tbl2
file_path_tbl2 = '../files/input/tbl2.csv'
data_tbl2 = pd.read_csv(file_path_tbl2, names=['K1', 'c21', 'c22', 'c23', 'c24', 'c25'], header=None)
```

```python
# Convertir la columna 'c23' a formato de fecha
data_tbl2['c23'] = pd.to_datetime(data_tbl2['c23'])

# Extraer el año de la columna 'c23'
data_tbl2['year'] = data_tbl2['c23'].dt.year

# Calcular el promedio de 'c21' por año
average_c21_per_year = data_tbl2.groupby('year')['c21'].mean()
average_c21_per_year
```

```python
# Encontrar el registro con el menor valor en la columna 'c21' de la tabla 'tbl2'
min_c21_record = data_tbl2[data_tbl2['c21'] == data_tbl2['c21'].min()]

min_c21_record
```

```python
# Filtrar los registros de la tabla 'tbl0' donde 'c02' es mayor o igual a 300
filtered_tbl0_c02 = data_tbl0[data_tbl0['c02'] >= 300]

filtered_tbl0_c02
```

```python
# Filtrar los registros de la tabla 'tbl1' para el año 2018
data['c14'] = pd.to_datetime(data['c14'])  # Asegurar que 'c14' esté en formato de fecha
total_records_2018 = data[data['c14'].dt.year == 2018].shape[0]

total_records_2018
```

```python
# Calcular tanto el valor máximo como el mínimo de la columna 'c12' para cada letra en la columna 'K0'
extreme_values_c12_per_k0 = data.groupby('K0')['c12'].agg(['max', 'min']).reset_index('K0')
extreme_values_c12_per_k0
```

```python
# Filtrar los registros donde 'c13' es mayor a 400 y luego calcular el valor promedio de 'c12' para cada letra en 'K0'
average_c12_per_k0_above_400 = data[data['c13'] > 400].groupby('K0')['c12'].mean()
average_c12_per_k0_above_400
```

```python
# Realizar la unión de las tablas tbl1 y tbl2 utilizando la columna 'K1' como clave
merged_data = pd.merge(data, data_tbl2, on='K1', suffixes=('_tbl1', '_tbl2'))

merged_data.head()
```

```python
# Calcular el promedio de 'c21' para cada letra en 'K0' de tbl1 con los registros filtrados
average_c21_per_k0 = merged_data.groupby('K0')['c21'].mean()
average_c21_per_k0
```

```python
import sqlite3

import pandas as pd  # type: ignore


def load_data():
    """Load data."""
    conn = sqlite3.connect(":memory:")
    cur = conn.cursor()

    with open("../tests/create_tables.sql", encoding="utf-8") as file:
        cur.executescript(file.read())

    return conn, cur

```

```python
"""Test 08."""
conn, _ = load_data()
with open("../homework/pregunta_08.sql", encoding="utf-8") as file:
    query = file.read()
result = pd.read_sql_query(query, conn)
result
```
